```python
#null values more than 80%
null_variables_eligibility_80=dict(((df.isnull().sum()/df.shape[0])*100 )>80)
```

```python
null_variables_80=[]
for x,y in null_variables_eligibility_80.items():
    if y==True:
        null_variables_80.append(x)
```

```python
#dropping the columns with more than 80% null values
df.drop(columns=null_variables_80 ,inplace=True)
```

```python
# given dataset contains categorical and numerical data
categorical_df=df.select_dtypes(exclude=np.number)
numerical_df=df.select_dtypes(include=np.number)
```

```python
categorical_df
```

```python
numerical_df
```

```python
#finding percentage of null values for each column
(df[categorical_df.columns].isnull().sum()/df[categorical_df.columns].shape[0])*100
```

```python
# finding categorical columns with null values more than 5%
filter=(df[categorical_df.columns].isnull().sum()/df[categorical_df.columns].shape[0])*100 >5
_to_delete=df[categorical_df.columns].loc[:,filter.values]
_delete_column=list(_to_delete.columns)
```

```python
_delete_column
```

```python
#checking whether the variables are relvant or not
df[_delete_column ]# not much relevance
```

```python
#dropping the identified columns
df.drop(columns=_delete_column,inplace=True)
```

```python
#dropping the identified columns from categorical_df
categorical_df.drop(columns=_delete_column,inplace=True)
```

```python
df.shape#4 columns deleted
```

```python
#replacing null values in categorical columns with mode
for cols in categorical_df.columns:
    df[cols].fillna(df[cols].mode()[0],inplace=True)
```

```python
#checking again for null values 
(df[categorical_df.columns].isnull().sum()/df[categorical_df.columns].shape[0])*100 
```

```python
#finding the percentage of null values in numerical columns
(df[numerical_df.columns].isnull().sum()/df[numerical_df.columns].shape[0])*100
```

```python
#replacing null values in numerical columns with mean
for cols in numerical_df.columns:
    df[cols].fillna(df[cols].mean(),inplace=True)
```

```python
#checking again for null values
(df[numerical_df.columns].isnull().sum()/df[numerical_df.columns].shape[0])*100
```
